# Bottlenecks of varying strength

```python
from msp_sim import *
from tree_LD_analysis import *
import numpy as np 
import msprime as msprime 
import pandas as pd
import tskit
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
btl_sim=Bottleneck(T=100,s=0.8,Ne=20_000) # initialise a simulation with constant size
btl_sim.mutation_rate = 1e-5
num_reps=1_000
```

```python
exps=range(100)
btl_times=[2*(10**(e/10)) for e in exps]
btl_results_t=topology_bottleneck(btl_sim, # simulate and mutate!
                          num_replicates=num_reps, # 1,000 replicates in each scenario
                          btl_times=btl_times,
                          btl_size=0.8,
                          plot_dems=True,signed=True) 
```

```python
for i in range(len(btl_times)):
    print(np.sum(np.isnan(btl_results_t.r2[:,i,0])))
```

```python
for i in range(len(btl_times)):
    plt.errorbar(btl_times[i],
                 np.nanmean(btl_results_t.r2[:,i,0]**2),
                 # for clarity I'm plotting error bars which are 0.1 st dev. 
                 np.sqrt(np.nanvar(btl_results_t.r2[:,i,0]**2)),marker='o', 
                 mfc='tomato',
                 mec="lightsteelblue",
                 c="lightsteelblue")
plt.ylabel("$r^2$")
plt.xlabel("Time of bottleneck");
plt.xscale("log")
plt.savefig('btl1.pdf')
```

```python
for i in range(len(btl_times)):
    plt.scatter([btl_times[i]]*4,
                 np.nanmean(btl_results_t.tops[:,i,:],axis=0),
                 marker='o',c=[1,2,3,4])
plt.ylabel("mean percentage of trees")
plt.xlabel("Time of bottleneck");
plt.xscale("log")
plt.savefig('btl2.pdf')
```

```python
classes=["above","other side","same branch","opposite branch"]
for i in range(3):
    plt.scatter([btl_times],
                 np.nanmean(btl_results_t.tops[:,:,i],axis=0),
                 marker='o',label=classes[i])
plt.ylabel("mean percentage of trees")
plt.xlabel("Time of bottleneck")
plt.xscale("log")
plt.legend();
plt.savefig('btl4.pdf')
```

```python
for i in range(len(btl_times)):
    plt.errorbar(btl_times[i],
                 np.nanmean(btl_results_t.fracs[:,i]),
                 np.nanvar(btl_results_t.fracs[:,i]),marker='o',
                 mfc='tomato',
                 mec="lightsteelblue",
                 c="lightsteelblue")
plt.ylabel("fraction of tree made up of internal branches")
plt.xlabel("Time of bottleneck");
plt.xscale("log")
plt.savefig('btl3.pdf')
```

```python
num_reps=100
exps=range(50)
btl_sizes=[e/50 for e in exps]
btl_results_s=topology_bottleneck_size(btl_sim, # simulate and mutate!
                          num_replicates=num_reps, # 1,000 replicates in each scenario
                          btl_time=200,
                          btl_sizes=btl_sizes,
                          plot_dems=True,signed=True) 
```

```python
for i in range(len(btl_sizes)):
    print(np.sum(np.isnan(btl_results_s.r2[:,i,0])))
```

```python
for i in range(len(btl_sizes)):
    plt.errorbar(btl_sizes[i],
                 np.nanmean(btl_results_s.r2[:,i,0]**2),
                 # for clarity I'm plotting error bars which are 0.1 st dev. 
                 np.sqrt(np.nanvar(btl_results_s.r2[:,i,0]**2)),marker='o', 
                 mfc='tomato',
                 mec="lightsteelblue",
                 c="lightsteelblue")
plt.ylabel("$r^2$")
plt.xlabel("Proportion of lineages made to coalesce");

plt.savefig('btl21.pdf')
```

```python
for i in range(len(btl_sizes)):
    plt.scatter([btl_sizes[i]]*4,
                 np.nanmean(btl_results_s.tops[:,i,:],axis=0),
                 marker='o',c=[1,2,3,4])
plt.ylabel("mean percentage of trees")
plt.xlabel("Proportion of lineages made to coalesce");

plt.savefig('btl22.pdf')
```

```python
classes=["above","other side","same branch","opposite branch"]
for i in range(3):
    plt.scatter([btl_sizes],
                 np.nanmean(btl_results_s.tops[:,:,i],axis=0),
                 marker='o',label=classes[i])
plt.ylabel("mean percentage of trees")
plt.xlabel("Proportion of lineages made to coalesce");

plt.legend();
plt.savefig('btl24.pdf')
```

```python
for i in range(len(btl_sizes)):
    plt.errorbar(btl_sizes[i],
                 np.nanmean(btl_results_s.fracs[:,i]),
                 np.nanvar(btl_results_s.fracs[:,i]),marker='o',
                 mfc='tomato',
                 mec="lightsteelblue",
                 c="lightsteelblue")
plt.ylabel("fraction of tree made up of internal branches")
plt.xlabel("Proportion of lineages made to coalesce");
plt.savefig('btl23.pdf')
```
